```python
def dataset_filler(dataset):
    rows, columns = dataset.shape
    i = 0
    j = 0
    while i < rows:
        j = 0
        while j < columns:
            regex = re.compile(r'\b{0}\b'.format(dataset.columns[j]))
            dataset.iloc[i, j] = len(re.findall(regex, dataset.index[i]))
            j += 1
        i += 1
    return(dataset)
```

```python
num_of_words = dataset_filler(dataset)
num_of_words.to_excel('sample.xlsx')
```
